## Using the Linear Regression Model for average September SIE


- Using an "ice-free" threshold of mean September SIE <= 1 million km^2
- Investigating how the choice of training data cutoff year influences the final prediction

```python
monthly_df = create_SIE_df("monthly")
sept_df = monthly_df[monthly_df["Month"] == 9].drop(columns=["Month"])

future_years = list(range(2025, 2300))
X_train, y_train, _,_ = train_test_split(sept_df, year_split=2011)

first_year,_ = linear_regression_year_prediction(X_train, y_train, future_years)
save_results_to_csv_year(first_year, "Linear Regression", "Year", "September mean SIE", "1978-2011")
```
